Import Libraries & Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel('supermarket_sales.xlsx')
```

Create a basic Violin plot on one NUMERIC variable e.g. Total

```python
sns.violinplot(data=df,x='Total')
plt.show()
```

Create a VIOLINPLOT on two categorical and one numeric variable and use split

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=df,x='Product line',y='Total')
plt.show()
```

Change the box in the VIOLINPLOT to horzontal lines

```python
plt.figure(figsize=(5,15))

sns.violinplot(data=df,y='Product line',x='Total')
plt.show()
```

Draw line for each observations in a VIOLINPLOT

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=df,x='Product line',y='Total',palette="muted", split=True)
plt.show()

```

Change the amount of smoothing using bw attribut

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=df,x='Product line',y='Total',palette="muted", split=True,bw=0.2)
plt.show()

```

Cut out the extreme values

```python
plt.figure(figsize=(15,5))
sns.violinplot(data=df,x='Product line',y='Total',palette="muted", split=True,bw=0.2,cut=2)
plt.show()

```
