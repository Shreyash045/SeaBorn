IMPORT DATA SET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('supermarket_sales.xlsx')
df.head()
```

Create a basic KDE plot for [Total] column

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , x='Total')
plt.show()
```

 Create KDE for all the numeric variables in dataframe

## *   List item
*   List item



```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , multiple='layer')
plt.show()
```

Adjust the smooting using bw_adjust

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , x='Unit price',bw_adjust=0.6)
plt.show()
```

Group the KDE on a category variable [Payment]

```python
plt.figure(figsize=(15,5))
df['Payment']=df['Payment'].replace({'Ewallet':0,'Cash':1,'Credit card':2})
sns.kdeplot(data=df , x='Total',y='Payment')


plt.show()
```

Stack KDE on a category using MULTIPLE arguement

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , x='Unit price',multiple='layer')
plt.show()
```

Use log scaling to map the variable in KDE [log_scale=True]

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , x='Unit price',log_scale=True)
plt.show()
```

Change styling of hued KDE using linewidth, palette, alpha etc...

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , x='Unit price',multiple='layer',linewidth=0.5,palette="warm")
plt.show()
```

Group the bivariate KDE on a categorical variable and show the contours [ x=unit price , y = gross income hue =gender ]

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df , x='Unit price',y='gross income',hue='Gender',multiple='layer')
plt.show()
```
