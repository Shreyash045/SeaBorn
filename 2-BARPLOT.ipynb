Display the SEBORN dataset names and load one of them for example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Import Sales Dataset 

```python
df = pd.read_excel('supermarket_sales.xlsx')
df.head()
```

Create Basic BarPlot Using Product line and Total


```python
plt.figure(figsize=(15,8))
sns.barplot(x='Product line',y='Total',data=df)
```

Add hue to get the count on two categories i.e. Product line and Gender

```python
plt.figure(figsize=(15,8))
sns.barplot(x='Product line',y='Total',data=df,hue='Gender')
```

Make the barplot HORIZONTAL

```python
plt.figure(figsize=(15,8))
sns.barplot(y='Product line',x='Total',data=df,hue='Gender')
```

Plot the bars in a alphabetical Order of Product line

```python
plt.figure(figsize=(15,8))
df.sort_values('Product line')
sns.barplot(x='Product line',y='Total',data=df,hue='Gender')
```

Add CAP on the ERROR BAR [capsize=0.2]

```python
plt.figure(figsize=(15,8))
df.sort_values('Product line')
sns.barplot(x='Product line',y='Total',data=df,hue='Gender',capsize=0.)
```

REMOVE the error bar using ci

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=df,hue='Gender',capsize=0.2,ci=68)
plt.show()
```

Change bar colors using COLOR attribute

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=df,hue='Gender',capsize=0.2,ci=100,color="magenta")
plt.show()
```

Change color using PALLETE and Saturation attributes

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=df,hue='Gender',capsize=0.2,ci=68,color="magenta",palette="Blues_d")
plt.show()
```

CHANGE DEFAULT AGGREGATION METHOD USING ESTIMATOR PARAMETER

```python
from numpy import median
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=df,hue='Gender',capsize=0.2,ci=68,color="magenta",estimator=median)
plt.show()
```
