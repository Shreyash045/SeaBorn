IMPORT DATA SET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('supermarket_sales.xlsx')
```

 Create a basic RUG plot for one variable - Unit Price

```python
sns.rugplot(x='Total',data=df)
```

Create a RUG plot for two variables - Unit Price , gross income

```python
sns.rugplot(x='Unit price',y='gross income',data=df)
```

Group it by a categorical variable using hue - Gender

```python
sns.rugplot(x='Unit price',y='gross income',data=df,hue='Gender')
```

Change the height of Rugs in the RUG Plot

 Combine this with a KDE Plot

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df, x='Unit price',y='gross income',hue='Gender',fill=True)
sns.rugplot(data=df, x='Unit price',y='gross income',hue='Gender')
```

Combine this with a SCATTER PLOT

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=df, x='Unit price',y='gross income',hue='Gender')
sns.rugplot(data=df, x='Unit price',y='gross income',hue='Gender')
```

Show Rugs outside of the Axis

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=df, x='Unit price',y='gross income',hue='Gender',fill=True)
sns.rugplot(data=df, x='Unit price',y='gross income',hue='Gender',height=-0.01,clip_on=False)
```
