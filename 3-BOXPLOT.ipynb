Import Libraries & Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('supermarket_sales.xlsx')
df.head()
```

Create a basic BOX plot on one NUMERIC variable e.g. Total

```python
sns.boxplot(x="Total",data=df)
plt.show()
```

Create a basic BOX plot on one numeric variable by a CATEGORICAL variable like Payment 

```python
sns.boxplot(x='Payment',y='Total',data=df)
plt.show()
```

Create a basic BOX plot on one numeric variable by TWO CATEGORICAL variable using HUE attribute

```python
sns.boxplot(x='Payment',y='Total',data=df,hue='Gender')
plt.legend(loc='upper center')
plt.show()
```

 Add MEAN marker in the box plot using Showmeans attribute and change its style using meanprops
 meanprops={"marker":"o","markerfacecolor":"white","markersize":"10","markeredgecolor":"black"}

```python
meanprops={"marker":"o","markerfacecolor":"white","markersize":"10","markeredgecolor":"black"}
plt.figure(figsize=(15,8))
sns.boxplot(x='Payment',y='Total',data=df,hue='Gender',showmeans=meanprops)
plt.legend(loc='upper center')
plt.show()
```

 Make HORIZONTAL box plot

```python
meanprops={"marker":"o","markerfacecolor":"white","markersize":"10","markeredgecolor":"black"}
plt.figure(figsize=(15,8))
sns.boxplot(y='Payment',x='Total',data=df,hue='Gender',showmeans=meanprops)
plt.legend(loc='upper center')
plt.show()
```

Change PALETTE, LINE WIDTH etc..

```python
meanprops={"marker":"o","markerfacecolor":"white","markersize":"10","markeredgecolor":"black"}
plt.figure(figsize=(15,8))
sns.boxplot(x='Payment',y='Total',data=df,hue='Gender',showmeans=meanprops,palette='cool',linewidth=0.9)
plt.legend(loc='upper center')
plt.show()
```
