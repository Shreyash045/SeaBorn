Load Seaborn Flights DataSet and Create Pivots On Month , Year , Passengers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts

df = sns.load_dataset('flights')
flights = df.pivot("month", "year", "passengers")
print(flights)

```

Create HeatMap with linewidth = 0.5 , and cmap = summer

```python
sns.heatmap(flights,linewidth=0.5,cmap="spring")
```

Create HeatMap Using Car_Crashes Dataset with corr() function
Set Properties annot = True , linewidth = 0.5 , cmap = YlGnBu

```python
car = sns.load_dataset ('car_crashes')
df = car.corr()
sns.heatmap(df , linewidth = 0.5 , cmap = 'YlGnBu',annot=True)
```

Create ClusterMap of flights Dataset with linewidth = 0.5 , and cmap = coolwarm

```python
sns.clustermap(flights , linewidth=0.5 , cmap = 'winter')
```

Modify above cluster for rows Only

```python
sns.clustermap(flights , linewidth=0.5 , cmap = 'spring',col_cluster=False)
```

Create FacetGridPlot On Tips Dataset On column time and Display ScatterPlot of total_bill and tip Using map function


```python
tips = sns.load_dataset('tips')
f = sns.FacetGrid ( tips , col = 'time' )
f.map ( sns.scatterplot ,'total_bill','tip')
```

Create FacetGridPlot On Tips Dataset On row sex and Display ScatterPlot of total_bill and tip Using map function

```python
f = sns.FacetGrid ( tips , row='sex' )
f.map ( sns.scatterplot , 'total_bill','tip')
```
