Read SuperMarketSales Data Set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('supermarket_sales.xlsx')
df.head()
```

Create a basic COUNTPLOT to get number of transaction for each of the product line and change the figure size

```python
sns.countplot(x='Product line',data=df)
plt.show()
```

Make it horizontal bar plot

```python
sns.countplot(y='Product line',data=df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='Product line',data=df,hue='Gender')
plt.show()
```

Using different color palette

```python
plt.figure(figsize=(15,10))
sns.countplot(x='Product line',data=df,hue='Gender',palette='cool')
plt.show()
```

Change Style using facecolor, linewidth and edge color

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line',data=df,facecolor = (0,0,0,0) ,linewidth=5 ,edgecolor=sns.color_palette('dark',3))
```
