Import Libraries & Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('supermarket_sales.xlsx')
df.head()
```

Create a basic SWARM PLOT on one NUMERIC variable e.g. Total

```python
sns.swarmplot(data=df,y='Total')
plt.show()
```

Create SWARM PLOT group by CATEGORIES

```python
plt.figure(figsize=(15,5))
sns.swarmplot(data=df,y='Total',x='Payment')
plt.show()
```

Showing Hues Separatly on the categorical axis

```python
plt.figure(figsize=(15,5))
sns.swarmplot(data=df,y='Total',x='Payment',hue='Gender')
plt.show()
```

Styling a SWARM - Change the marker, size, color, edge color....

```python
plt.figure(figsize=(15,5))
sns.swarmplot(data=df,y='Total',x='Payment',hue='Gender',size=5,color="magenta",edgecolor="black")
plt.show()
```

Overlay a SWARM PLOT on a VIOLIN PLOT

```python
plt.figure(figsize=(15,5))
sns.swarmplot(data=df,y='Total',x='Payment',hue='Gender')
sns.violinplot(data=df,y='Total',x='Payment',hue='Gender')

plt.show()
```

Overlay a SWARM PLOT on a BOX PLOT

```python
plt.figure(figsize=(15,5))
sns.swarmplot(data=df,y='Total',x='Payment',hue='Gender')
sns.boxplot(data=df,y='Total',x='Payment',hue='Gender')

plt.show()
```
